```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("supd.csv")

avg_app_usage_count = df.groupby("Age")["App_Usage_Count"].mean().reset_index()

plt.figure()
plt.scatter(avg_app_usage_count["Age"],
avg_app_usage_count["App_Usage_Count"])
plt.xlabel("Age")
plt.ylabel("Average App Usage Count")
plt.title("Age vs Average App Usage Count")
plt.show()
```
