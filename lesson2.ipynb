```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("supd.csv")

avg_app_usage_count = df.groupby("Age")["App_Usage_Count"].mean().reset_index()

plt.figure()
plt.scatter(avg_app_usage_count["Age"],
avg_app_usage_count["App_Usage_Count"])
plt.xlabel("Age")
plt.ylabel("Average App Usage Count")
plt.title("Age vs Average App Usage Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("supd.csv")
df.groupby("Age")[["Daily_Phone_Hours","Social_Media_Hours","Weekend_Screen_Time_Hours","Sleep_Hours"]].mean().plot(xlabel="Age", ylabel="Average Hours",title="Average Screen Time and Sleep by Age")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load dataset
df = pd.read_csv("supd.csv")
# Group by Age and calculate mean
grouped = df.groupby("Age")[[
 "Daily_Phone_Hours",
 "Social_Media_Hours",
 "Weekend_Screen_Time_Hours",
 "Sleep_Hours"
]].mean().reset_index()
# Plot
plt.figure()
plt.plot(grouped["Age"], grouped["Daily_Phone_Hours"], label="Daily Phone Hours")
plt.plot(grouped["Age"], grouped["Social_Media_Hours"], label="Social Media Hours")
plt.plot(grouped["Age"], grouped["Weekend_Screen_Time_Hours"], label="Weekend Screen Time Hours")
plt.plot(grouped["Age"], grouped["Sleep_Hours"], label="Sleep Hours")
plt.xlabel("Age")
plt.ylabel("Average Hours")
plt.title("Average Screen Time and Sleep by Age")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load dataset
df = pd.read_csv("supd.csv") # change path if needed
# Calculate total hours
daily_phone_total = df['Daily_Phone_Hours'].sum()
social_media_total = df['Social_Media_Hours'].sum()
Weekend_Hours = df['Weekend_Screen_Time_Hours'].sum()
# Data for pie chart
labels = ['Daily Phone Hours', 'Social Media Hours', 'Weekend Screen Time Hours']
sizes = [daily_phone_total, social_media_total, Weekend_Hours]
# Plot pie chart
plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Daily Phone vs Social Media Usage vs Weekend Screen Time Hours')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("supd.csv")
ages = [18, 20, 25, 30, 35, 40, 45, 50]

# Filter dataset
filtered = df[df["Age"].isin(ages)]

# Group by Age and calculate average productivity
age_productivity = filtered.groupby("Age")["Work_Productivity_Score"].mean().reindex(ages)
# Group by Age and calculate average productivity hours
#age_productivity = df.groupby("Age")["Work_Productivity_Score"].mean()

# Plot bar chart
plt.figure()
age_productivity.plot(kind="bar")

plt.xlabel("Age")
plt.ylabel("Average Productivity Score")
plt.title("Average Productivity Score by Age")

plt.show()
```

```python
# histogram
import pandas as pd
import matplotlib.pyplot as plt
# Load dataset
df = pd.read_csv("supd.csv")
plt.figure()
plt.hist(df["Weekend_Screen_Time_Hours"], bins=10)
plt.xlabel("Weekend Screen Time Hours")
plt.ylabel("Frequency")
plt.title("Distribution of Weekend Screen Time Hours")
plt.show()
```

```python
# density
import pandas as pd
import matplotlib.pyplot as plt
# Load dataset
df = pd.read_csv("supd.csv")
# Density chart
df["Weekend_Screen_Time_Hours"].plot(kind="kde")
plt.xlabel("Weekend Screen Time Hours")
plt.ylabel("Density")
plt.title("Density Chart of Weekend Screen Time Hours")
plt.show()
```

```python
# histogram + density
import pandas as pd
import matplotlib.pyplot as plt
# Load dataset
df = pd.read_csv("supd.csv")
plt.figure()
# Histogram (normalized so it matches density scale)
df["Weekend_Screen_Time_Hours"].plot(kind="hist", bins=10, density=True)
# Density curve
df["Weekend_Screen_Time_Hours"].plot(kind="kde")
plt.xlabel("Weekend Screen Time Hours")
plt.ylabel("Density")
plt.title("Histogram and Density of Weekend Screen Time Hours")
plt.show()
```
