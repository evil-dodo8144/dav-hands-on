```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([2,4,1,3,5])

# r for red and - for solid line and s for square markers
plt.plot(x,y,'g-s')
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.title("Graph with red square and headings")
plt.show()
```

```python
## bar chart
import matplotlib.pyplot as plt
import numpy as np
departments = ["cse","ece","ee","ce"]
students=[120,50,90,20]
plt.bar(departments,students)
plt.xlabel("departments")
plt.ylabel("students")
plt.title("students per Department")
plt.show()
```
